```python
# All modules import for use
# data manipulate and viszulize 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing Data for cleaning 
wine = pd.read_csv('winequality-red.csv')
```

```python
wine.describe()
```

```python
wine.isnull().value_counts()
```

```python
wine.dtypes
```

```python
# Plot the distribution of wine quality
plt.figure(figsize=(10, 6))
sns.countplot(x='quality', data=wine, palette='viridis')
plt.title('Distribution of Wine Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
# List of features to plot
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 
            'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

# Plot histograms for each feature
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features):
    plt.subplot(4, 3, i+1)
    sns.histplot(wine[feature], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='alcohol', y='quality', data=wine, hue='quality', palette='viridis', s=100)
plt.title('Scatter Plot of Wine Quality vs. Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = wine.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```
